### Gaussian vs. SP

```python
dt_cecs = pd.read_csv('results_SPeda_Funcs_rec.csv', index_col = 0)
pd.set_option('display.max_rows', dt_cecs.shape[0]+1)
dt_cecs_gb_mean = dt_cecs[dt_cecs['alpha'] == 0.4].groupby(['cost_function', 'size_gen', 'l']).mean().drop('alpha', axis=1)
dt_cecs_gb_std  = dt_cecs[dt_cecs['alpha'] == 0.4].groupby(['cost_function', 'size_gen', 'l']).std().drop('alpha', axis=1)
dt_cecs_gb = dt_cecs_gb_mean.reset_index().merge(dt_cecs_gb_std.reset_index(), on=['cost_function', 'l', 'size_gen'],
                                             suffixes = ['_mean', '_std'])
# dt_cecs_gb
```

```python
dt_cecs_g = pd.read_csv('results_Geda_Funcs_rec.csv', index_col = 0)
pd.set_option('display.max_rows', dt_cecs_g.shape[0]+1)
dt_cecs_g_gb_mean = dt_cecs_g[dt_cecs_g['alpha'] == 0.4].groupby(['cost_function', 'size_gen', 'l']).mean().drop(['alpha', 'it'], axis=1)
dt_cecs_g_gb_std  = dt_cecs_g[dt_cecs_g['alpha'] == 0.4].groupby(['cost_function', 'size_gen', 'l']).std().drop(['alpha', 'it'], axis=1)
dt_cecs_g_gb = dt_cecs_g_gb_mean.reset_index().merge(dt_cecs_g_gb_std.reset_index(), on=['cost_function', 'l', 'size_gen'],
                                             suffixes = ['_mean', '_std'])
# dt_cecs_g_gb
```

```python
dt_comp_g_sp = dt_cecs_gb.merge(dt_cecs_g_gb, on=['cost_function', 'l', 'size_gen'], suffixes = ['_a', '_b'])
dt_comp_g_sp.columns = ['cost_function', 'N', 'l', 'mu_speda', 'std_speda', 'mu_egna', 'std_egna']
dt_comp_g_sp = dt_comp_g_sp.groupby(['cost_function', 'l', 'N']).mean().round(2)
print(dt_comp_g_sp.to_latex())
```

```python
dt_sp_vs_g = dt_cecs_gb_mean.reset_index().merge(dt_cecs_g_gb_mean.reset_index(), on=['cost_function', 'l', 'size_gen'],
                                             suffixes = ['_sp', '_g']).round(2) # .drop(['alpha_g', 'alpha_sp', 'it'], axis=1)
# dt_sp_vs_g

dt_sp_vs_g_std = dt_cecs_gb_std.reset_index().merge(dt_cecs_g_gb_std.reset_index(), on=['cost_function', 'l', 'size_gen'],
                                             suffixes = ['_sp', '_g']).round(2)

dt_sp_vs_g.columns = ['cost_function', 'N', 'l', 'SPEDA', 'EGNA']
```

```python
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec1', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec1']['SPEDA'] - 100
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec2', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec2']['SPEDA'] - 200
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec4', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec4']['SPEDA'] - 400
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec5', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec5']['SPEDA'] - 500
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec6', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec6']['SPEDA'] - 600
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec7', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec7']['SPEDA'] - 700
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec8', 'SPEDA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec8']['SPEDA'] - 900

dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec1', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec1']['EGNA'] - 100
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec2', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec2']['EGNA'] - 200
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec4', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec4']['EGNA'] - 400
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec5', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec5']['EGNA'] - 500
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec6', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec6']['EGNA'] - 600
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec7', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec7']['EGNA'] - 700
dt_sp_vs_g.loc[dt_sp_vs_g['cost_function'] == 'cec8', 'EGNA'] = dt_sp_vs_g[dt_sp_vs_g['cost_function'] == 'cec8']['EGNA'] - 900
```

```python
from functools import reduce

df1 = dt_sp_vs_g[(dt_sp_vs_g['l'] == 10) & (dt_sp_vs_g['N'] == 200)].drop(['N', 'l'], axis=1)
df2 = dt_sp_vs_g[(dt_sp_vs_g['l'] == 10) & (dt_sp_vs_g['N'] == 300)].drop(['N', 'l'], axis=1)
df3 = dt_sp_vs_g[(dt_sp_vs_g['l'] == 15) & (dt_sp_vs_g['N'] == 200)].drop(['N', 'l'], axis=1)
df4 = dt_sp_vs_g[(dt_sp_vs_g['l'] == 15) & (dt_sp_vs_g['N'] == 300)].drop(['N', 'l'], axis=1)
dfs = [df1, df2, df3, df4]

df_final = reduce(lambda left, right: pd.merge(left,right, on=['cost_function'], suffixes = ['_a', '_b']), dfs)
df_final.columns = ['cost_function', 'SPEDA_1', 'EGNA_1', 'SPEDA_2', 'EGNA_2', 'SPEDA_3', 'EGNA_3', 'SPEDA_4', 'EGNA_4']
df_final = df_final[['cost_function', 'SPEDA_1', 'SPEDA_2', 'SPEDA_3', 'SPEDA_4','EGNA_1', 'EGNA_2', 'EGNA_3', 'EGNA_4']]
```

```python
import pandas as pd


def load_data(algorithm, dim, sep= ','):
    dt = pd.read_csv(str(dim) + 'd_' + str(algorithm) + '.csv', sep=sep)
    for i in range(1, 20 + 1):
        dt.loc[dt['cost_function'] == 'cec' + str(i), 'result'] = round(dt[dt['cost_function'] == 'cec' + str(i)]['result'] - i*100, 2)
    # dt.loc[dt['cost_function'] == 'cec8', 'result'] = round(dt[dt['cost_function'] == 'cec8']['result'] - 100, 2)
    # dt = dt.groupby(variables).mean().round(2)
    return dt

def loads():
    dt_egna_30d = load_data('egna', 30).round(2)
    dt_egna_30d = dt_egna_30d[(dt_egna_30d['size_gen'] == 300) & (dt_egna_30d['alpha'] == 0.6) & (dt_egna_30d['l'] == 10)]
    dt_egna_30d = dt_egna_30d[['it', 'result', 'cost_function']]
    dt_egna_30d['algorithm'] = 'EGNA'

    dt_jade_30d = load_data('jade_', 30, sep=' ').round(2)
    dt_jade_30d = dt_jade_30d[['it', 'result', 'cost_function']]
    dt_jade_30d['algorithm'] = 'JADE'

    dt_cmaes_30d = load_data('cmaes_prueba', 30).round(2)
    dt_cmaes_30d = dt_cmaes_30d[['it', 'result', 'cost_function']]
    dt_cmaes_30d['algorithm'] = 'CMAES'

    dt_speda_30d = load_data('speda', 30).round(2)
    dt_speda_30d = dt_speda_30d[(dt_speda_30d['size_gen'] == 300) & (dt_speda_30d['alpha'] == 0.4) & (dt_speda_30d['l'] == 15)]
    dt_speda_30d.columns = ['it', 'alpha', 'result', 'size_gen', 'cost_function', 'l']
    dt_speda_30d = dt_speda_30d[['it', 'result', 'cost_function']]
    dt_speda_30d['algorithm'] = 'SPEDA'
    # dt_speda_30d['it'] = list(range(len(dt_speda_30d)))

    dt_emna_30d = load_data('emna', 30).round(2)
    dt_emna_30d = dt_emna_30d[(dt_emna_30d['size_gen'] == 300) & (dt_emna_30d['alpha'] == 0.4)]
    dt_emna_30d = dt_emna_30d[['it', 'result', 'cost_function']]
    dt_emna_30d['algorithm'] = 'EMNA'
    
    return dt_egna_30d, dt_jade_30d, dt_cmaes_30d, dt_speda_30d, dt_emna_30d

# datas = [dt_egna_30d, dt_jade_30d, dt_cmaes_30d, dt_speda_30d, dt_emna_30d]


# gb.to_csv('comparison_30d.csv')
# dt_tot = dt_egna_30d[:min_len].append(dt_jade_30d[:min_len]).append(dt_cmaes_30d[:min_len]).append(dt_speda_30d[:min_len]).append(dt_emna_30d[:min_len])

# datas = [dt_egna_30d, dt_jade_30d, dt_cmaes_30d, dt_speda_30d, dt_emna_30d]
# ceces = dt_egna_30d.cost_function.unique()
```
