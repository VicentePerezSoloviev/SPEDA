```python
def build_mesh_grid(dataset, variable, x, z):
    matrix = np.zeros((len(z), len(x)))
    # print(matrix.shape)
    for i in range(len(z)):
        # print(dataset.loc[z[i]][variable].round(2))
        matrix[i, :] = dataset.loc[z[i]][variable].round(2).values
    return matrix
```
