```python
dt = pd.read_csv('results_eda_30.csv', index_col=0)
gb_eda = dt.groupby(['alpha', 'size_gen']).mean().reset_index()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = gb_eda['alpha']
y = gb_eda['size_gen']
z = gb_eda['result']

fig = plt.figure(figsize=(6, 6))
plt.scatter(x, y,
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 200,
           c=z)
plt.show()
```
